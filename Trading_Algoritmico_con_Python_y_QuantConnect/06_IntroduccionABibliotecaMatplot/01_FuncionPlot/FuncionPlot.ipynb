Matplotlib es una biblioteca para la generación de gráficos a partir de datosa contenidos en listas.

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline # Guarda cada gráfica generada
```

```python
import numpy as np
```

```python
x = np.linspace(0,100,101)
```

```python
x
```

```python
y = np.sqrt(x) + 5
```

```python
y
```

```python
plt.plot(x,y, color = 'blue')
plt.xlabel('Eje X')
plt.ylabel("Eje Y")
plt.title("Titulo del gráfico")
plt.show() # Equivalente a print para gráficos
```
