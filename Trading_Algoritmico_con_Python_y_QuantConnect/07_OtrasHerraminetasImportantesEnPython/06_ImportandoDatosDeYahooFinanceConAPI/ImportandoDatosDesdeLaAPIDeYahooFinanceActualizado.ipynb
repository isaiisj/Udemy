```python
# ============================================
# 2. IMPORTAR LIBRERÍAS
# ============================================
import yfinance as yf
from datetime import datetime
import pandas as pd
```

```python
# ============================================
# 3. DEFINIR TICKERS Y FECHAS
# ============================================
# Define los tickers que quieres consultar
# Nota: Los índices como ^IXIC no tienen 'Adj Close'

tickers = ['NFLX', 'TSLA']  

# Define las fechas de inicio y fin
inicio = datetime(2018, 9, 26)
fin = datetime(2023, 9, 26)
```

```python
# ============================================
# 4. DESCARGAR DATOS
# ============================================
# IMPORTANTE: Usar auto_adjust=False para obtener la columna 'Adj Close'

datos = yf.download(tickers, start=inicio, end=fin, auto_adjust=False)
```

```python
# ============================================
# 5. EXPLORAR LOS DATOS
# ============================================
# Ver la estructura de los datos

print("Forma del DataFrame (filas, columnas):")
print(datos.shape)

print("Columnas disponibles:")
print(datos.columns.levels[0].tolist())

print("Tickers descargados:")
print(datos.columns.levels[1].tolist())

# Ver las primeras filas
print("Primeras 5 filas:")
print(datos.head())

# Ver todas las filas (si quieres ver todo completo)
print("Todos los datos:")
print(datos)
```

```python
# ============================================
# 6. ACCEDER A DATOS ESPECÍFICOS
# ============================================
# Acceder al precio ajustado de todos los tickers
print("Precios ajustados (Adj Close):")
print(datos['Adj Close'].head())

# Acceder al precio ajustado de un ticker específico
print("Precio ajustado de NFLX:")
print(datos['Adj Close']['NFLX'].head())
```

```python
# ============================================
# 7. OBTENER PRIMER Y ÚLTIMO VALOR
# ============================================
# Obtener el primer y último valor de Adj Close para NFLX
primer_valor_nflx = datos['Adj Close']['NFLX'].iloc[0]
ultimo_valor_nflx = datos['Adj Close']['NFLX'].iloc[-1]

print(f"NFLX - Primer Adj Close: ${primer_valor_nflx:.2f}")
print(f"NFLX - Último Adj Close: ${ultimo_valor_nflx:.2f}")

# Calcular el rendimiento porcentual
cambio_porcentual_nflx = ((ultimo_valor_nflx - primer_valor_nflx) / primer_valor_nflx) * 100
print(f"NFLX - Cambio total: {cambio_porcentual_nflx:.2f}%")

# Lo mismo para TSLA
primer_valor_tsla = datos['Adj Close']['TSLA'].iloc[0]
ultimo_valor_tsla = datos['Adj Close']['TSLA'].iloc[-1]

print(f"TSLA - Primer Adj Close: ${primer_valor_tsla:.2f}")
print(f"TSLA - Último Adj Close: ${ultimo_valor_tsla:.2f}")

cambio_porcentual_tsla = ((ultimo_valor_tsla - primer_valor_tsla) / primer_valor_tsla) * 100
print(f"TSLA - Cambio total: {cambio_porcentual_tsla:.2f}%")
```

```python
# ============================================
# 8. DATOS CON DIFERENTES INTERVALOS
# ============================================
# Descargar datos mensuales
datos_mensuales = yf.download(tickers, start=inicio, end=fin, 
                               interval='1mo', auto_adjust=False)

print(f"Datos mensuales - Total de meses: {len(datos_mensuales)}")
print("Primeros datos mensuales:")
print(datos_mensuales['Adj Close'].head())
```

```python
# ============================================
# 9. MÉTODO ALTERNATIVO: USAR TICKER INDIVIDUAL
# ============================================
# Si prefieres trabajar con un solo ticker
nflx = yf.Ticker('NFLX')
nflx_data = nflx.history(start='2018-09-26', end='2023-09-26')

print("Datos de NFLX usando método .history():")
print(nflx_data.head())
print("Nota: Con .history(), la columna 'Close' ya viene ajustada por defecto")
```
